```python
import pandas as pd
cars=pd.read_csv('mtcars.csv')
```

```python
cars.head()
```

```python
X=cars.iloc[:,[1,2,4,6]].values
```

```python
df = pd.DataFrame(X,columns=["mpg","cyl","hp","wt"])
df.head()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nbrs=NearestNeighbors(n_neighbors=3)
```

```python
nbrs.fit(df)
```

```python
features=['mpg','cylinder','hp','weight']
array=[]
print("Enter the desired Specification")
for i in range(4):
    x=float(input('Enter the {} of the car:'.format(features[i])))
    array.append(x)
```

```python
array
```

```python
rec_cars=nbrs.kneighbors([array])
rec_cars
```
